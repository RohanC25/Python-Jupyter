```python
olympix.columns  # Get column names in from Table
```

```python
col = ['gold_medal', 'silver_medal', 'bronze_medal', 'total_medal'] # list of medal variables
```

## Count plot of variables gold medal, silver medal, bronze medal, total medal

```python
n=1
plt.subplots(figsize = (20,16)) # ploting subplot
for i in col:
    plt.subplot(3,2,n)  # subplot dimensions
    ax = plt.gca()    # axis view
    sns.set(font_scale = 1.5)  # scale
    sns.countplot(x = olympix[i]) # count plot
    ax.set_title('Distribution of %s'%(i), fontsize=23) # give tile
    n+=1
    plt.tight_layout()
plt.suptitle('Distribution of variables', y=1.03, fontsize=30)  # give title to whole plot
plt.show()  # show plot
```

##### The 1st plot shows More than 5000  Participants have not won a single medal
##### The 2nd plot shows more than 5000 participants have not wo single silver medal
##### The 3rd plot shows more than 5000 participants have not won single bronze medal
##### The 4th plot shows nearly 8000 participants have won 1 medal.

# Top 10 Countries with most gold medals

```python
# sort on count of gold medals
top_gold = olympix.groupby('country')['gold_medal'].sum().sort_values(ascending = False).head(10) 

# Create bar plot country vs gold_medal
gold = top_gold.plot(kind = 'bar', title = 'country Vs sum of gold_medal', figsize = (10,7)) 

# Add data labels to the bars
for index, value in enumerate(top_gold):
    plt.text(index, value, str(value), ha='center', va='bottom')

gold.set_xticklabels(top_gold.index)  #Set x axis labesl
gold.set_yticklabels([])  # Set y axis tick labels

# Set axis labels
gold.set_ylabel('Sum_gold_medal')
gold.set_xlabel('country')

# Remove grid line
plt.grid(False)
```
