### Que 2. Use Crime dataset from LMS

I) find the aggregations like all moments of business decisions for all columns,value counts.

```python
import numpy as np
```

```python
a = pd.read_csv('D:/ExcelR Excel assignment/Python Assignments/crime_data.csv')
```

```python
column_to_rename = 'Unnamed: 0'
```

```python
a = a.rename(columns={column_to_rename: 'State'})
```

```python
aggregations = a.describe()
```

```python
value_counts = {}
```

```python
for column in a.columns:
    value_counts[column] = a[column].value_counts()
```

```python
print("Aggregations - Moments of Business Decisions for All Columns:")
print(aggregations)
print("\n   Value Counts for All Columns:")
for column, counts in value_counts.items():
    print(f"\nColumn: {column}")
    print(counts)
```

### II) do the plottings like plottings like histogram, boxplot, scatterplot, barplot, piechart,dot chart.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(a['Murder'],edgecolor = 'Black')
plt.title("Murder Frequency vs Rate")
plt.xlabel("Murder Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x=a["Assault"])
plt.title("Box PLot of Assault")
plt.show()
```

```python
plt.scatter(a['UrbanPop'], a['Rape'])
plt.title("UrbanPop vs Rape")
plt.xlabel("UrbanPop")
plt.ylabel("Rape")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='UrbanPop', y='Assault', data = a , color = 'Red')
plt.title("Bar Plot of Assault by UrbanPop")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.pie(a["Rape"], labels = a['State'], autopct = '%1.2f%%')
plt.title('Rape by State')
plt.axis('equal')
plt.show()
```

```python
plt.plot(a['Murder'], 'ro', markersize=8)
plt.title("Dot Chart of Murder")
plt.xlabel('Index')
plt.ylabel('Murder')
plt.show()
```

## 4. Use Bank Dataset from LMS

A) change all the categorical columns into numerical by creating Dummies and using label encoder.

```python
c = pd.read_csv("D:/ExcelR Excel assignment/Python Assignments/bankfull.csv",delimiter = ';')
c
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_columns = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
c1_dummies = pd.get_dummies(c, columns=categorical_columns, drop_first=True)
```

```python
label_encoder = LabelEncoder()
```

```python
c1_dummies['y'] = label_encoder.fit_transform(c['y'])
```
